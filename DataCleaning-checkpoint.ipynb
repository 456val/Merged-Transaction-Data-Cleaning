### **Project Overview**
This project focuses on cleaning and transforming one-day supply transaction dataset that contains multiple product categories and amounts merged within single cells. The goal is to restructure the data into a clear and analyzable format, making it easier to identify spending patterns and transaction summaries.  Through this process, the project demonstrates the importance of data cleaning and organization in turning messy raw data into valuable insights.

### **Objective**

1. To clean and transform merged transaction data into a structured format.  
2. To align each product category with its correct transaction amount.  
3. To prepare the data for clear analysis and visualization of spending patterns.  


### **Techniques and Tools Used**

-  **Techniques:** Data Cleaning, Data Transformation, Data Visualization, Descriptive Statistics  
-  **Tools:** Python (Pandas, Matplotlib)


<span style='font-size:20px'>**Step 1: Import Libraries**</span>

```python
import pandas as pd
df=pd.read_csv("invoices.csv")
df
```

<span style='font-size:20px'>**Step 2: Data Transformation and Reshaping Code**</span>



```python

df['Category'] = df['Category'].str.split(' \| ')
df['Amount'] = df['Amount'].str.split(' \| ')

# Create a new DataFrame by exploding the "Category" and "Amount" columns
new_df = pd.DataFrame({
    'Order ID': df['Order ID'].explode(),
    'Category': df['Category'].explode(),
    'Amount': df['Amount'].explode()
})

# Reset the index to have continuous integers
new_df = new_df.reset_index(drop=True)

# Display the resulting DataFrame
print(new_df)
```

<span style='font-size:20px'>**Step 3: Data export.**</span>


```python
# file path for the new CSV file
output_csv_file = 'output_data.csv'

# Save the 'new_df' DataFrame to a CSV file
new_df.to_csv(output_csv_file, index=False)

print(f"Data saved to {output_csv_file}")
```

<span style='font-size:20px'>**Step 4:Data transformation and summarization.**</span>

```python

# Split the 'Amount' values into separate rows
new_df = new_df.assign(Amount=new_df['Amount'].str.split(r'(?<=[0-9])\s+(?=[A-Za-z])'))

# Convert the 'Amount' values to float
new_df['Amount'] = new_df['Amount'].apply(lambda x: [float(i) for i in x])

# split the 'Amount' column to separate rows
new_df = new_df.explode('Amount')

# Group the data by 'Category' and calculate the sum of 'Amount' and the count of transactions
category_summary = new_df.groupby('Category').agg({'Amount': 'sum', 'Category': 'count'})

# Rename the columns for clarity
category_summary = category_summary.rename(columns={'Amount': 'Total Amount Spent', 'Category': 'Transaction Count'})

# Reset the index to make 'Category' a regular column
category_summary = category_summary.reset_index()

# Display the category summary
print(category_summary)
```

<span style='font-size:20px'>**Step 5: Category Summary Export Code:**</span>



```python

# the file path for the updated CSV file
output_csv_file = 'output_data.csv'

# Save the 'category_summary' DataFrame to the CSV file, overwriting the existing file
category_summary.to_csv(output_csv_file, index=False)
```
